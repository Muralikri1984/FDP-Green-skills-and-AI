```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import joblib
```

```python
df=pd.read_csv("solar_power_output.csv")
df
```

```python
x=df[['temperature','humidity','solar_irradiance','wind_speed']]
y=df['solar_power_output']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred,color='blue',label='Predicated vs Actual')
plt.plot([y.min(),y.max()],[y.min(),y.max()],color='red',linestyle='--',label='Perfect Fit')
plt.xlabel("Actual Solar Power Output")
plt.ylabel("Predicated Solar Power Output")
plttitle("Actual Vs Predicated solar output")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(len(y_test)),y_test,label="Actual Solar Power output",marker='o')
plt.plot(range(len(y_test)),y_pred,label="Predicated solar power output",marker='x')
plt.xlabel("Data Point Index")
plt.ylabel("Solar Power Output")
plt.title("Actual vs Predicated Solar output(line plot)")
plt.legend()
plt.show()
```

```python
joblib.dump(model,'solar_power_predication_model.pkl')
```
